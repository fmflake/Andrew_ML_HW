```python
def Sigmoid(z):
    g = 1/(1+np.exp(-z))
    return g
```

```python
df.insert(0, column='x0', value=1.0)
```

```python
df.head()
```

```python
X = np.array(df.iloc[:,:3]).T
y = np.array(df.iloc[:,3])
```

```python
theta = np.zeros(3)
```

```python
X.shape, theta.shape, y.shape
```

```python
num_iters= 1000
alpha = 0.00005
```

```python
def Computecost(X, y, theta):
    cost = 0.0
    m = len(y)
    error = theta.T@X-y
    cost = np.sum(error**2)/(2*m)
    return cost
```

```python
def ComputecostLogistic(X, y, theta):
    m = len(y)
    h = Sigmoid(theta.T@X)
    cost_term = -1*y*np.log(h)-(1-y)*np.log(1-h)
    return np.sum(cost_term/m)
```

```python
J = ComputecostLogistic(X, y, theta)
```

```python
J
```

```python
def gradientDescent(X, y, theta, alpha):
    m = len(y)
    term = (theta.T @ X - y) * X
    theta_new = theta - alpha/m*np.sum(term, axis=1)
    return theta_new
```

```python
def gradientDescentLogistic(X, y, theta, alpha):
    m = len(y)
    term = (Sigmoid(theta.T @ X) - y) * X
    theta_new = theta - alpha/m*np.sum(term, axis=1)
    return theta_new
```

```python
J_hist = np.zeros(num_iters)

for num in range(0, num_iters):
    theta_new = gradientDescentLogistic(X, y, theta, alpha)
    J = ComputecostLogistic(X, y, theta_new)
    theta = theta_new
    print("Iter:", num, " Cost:", J)
    J_hist[num] = J

print("theta=",theta)
```

```python
plt.plot(J_hist)
plt.xlabel('iteration')
plt.ylabel('cost')
```

```python
theta0_vals = np.linspace(-10,10,100)
theta1_vals = np.linspace(-1,4,100)
J_vals = np.zeros((len(theta0_vals),len(theta1_vals)))
```

```python
J_vals[1,1]
```

```python
for i in range (0,len(theta0_vals)):
    for j in range (0,len(theta1_vals)):
        t = np.array([theta0_vals[i], theta1_vals[j]])
        J_vals[i,j] = Computecost(X, y, t)
        
```

```python
P, Q = np.meshgrid(theta0_vals, theta1_vals, indexing='ij')
levels = np.logspace(-2, 3, 20)
```

```python
plt.contour(P, Q, J_vals, levels)
plt.xlabel("Theta_0")
plt.ylabel("Theta_1")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(P,Q,J_vals,cmap='coolwarm')
```
