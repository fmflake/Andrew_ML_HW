```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/ex1data1.txt', names = ["x1", "y"])
df.head()
```

```python
ax = df.plot(x='x1', y='y', kind='scatter', marker='x', c='red', lw=0.5)
ax.set_xlabel("Population of City in 10,000s")
ax.set_ylabel("Profit in $10,000s")
```

```python
df.insert(0, column='x0', value=1.0)
```

```python
X = np.array(df.iloc[:,:2]).T
y = np.array(df.iloc[:,2])
X[:,0],X[:,1],X[:,2]
len(X)
```

```python
theta = np.zeros(2)
```

```python
X.shape, theta.shape, y.shape
```

```python
num_iters= 1000
alpha = 0.01
```

```python
def Computecost(X, y, theta):
    cost = 0.0
    m = len(y)
    error = theta.T@X-y
    cost = np.sum(error**2)/(2*m)
    return cost
```

```python
J = Computecost(X, y, theta)
```

```python
J
```

```python
def gradientDescent(X, y, theta, alpha):
    m = len(y)
    term = (theta.T @ X - y) * X
    theta_new = theta - alpha/m*np.sum(term, axis=1)
    return theta_new
```

```python
J_hist = np.zeros(num_iters)

for num in range(0, num_iters):
    theta_new = gradientDescent(X, y, theta, alpha)
    J = Computecost(X, y, theta_new)
    theta = theta_new
    print("Iter:", num, " Cost:", J)
    J_hist[num] = J

print("theta=",theta)
```

```python
plt.plot(J_hist)
plt.xlabel('iteration')
plt.ylabel('cost')
```

```python
theta0_vals = np.linspace(-10,10,100)
theta1_vals = np.linspace(-1,4,100)
J_vals = np.zeros((len(theta0_vals),len(theta1_vals)))
```

```python
J_vals[1,1]
```

```python
for i in range (0,len(theta0_vals)):
    for j in range (0,len(theta1_vals)):
        t = np.array([theta0_vals[i], theta1_vals[j]])
        J_vals[i,j] = Computecost(X, y, t)
        
```

```python
P, Q = np.meshgrid(theta0_vals, theta1_vals, indexing='ij')
levels = np.logspace(-2, 3, 20)
```

```python
plt.contour(P, Q, J_vals, levels)
plt.xlabel("Theta_0")
plt.ylabel("Theta_1")
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(P,Q,J_vals,cmap='coolwarm')
```
